```python
datamap = {}
for vloc in video_loc:
    fvloc = os.path.join(rootloc,vloc)
    myfiles = os.listdir(fvloc)
    tloc = {} #mapping : number:filename
    vloc = {} #mapping : number:filename
    for fl in myfiles:
        if ".txt" in fl:
            name = fl.replace(".lines.txt","")
            tloc[name] = os.path.join(fvloc,fl)
        
        if ".jpg" in fl:
            name = fl.replace(".jpg","")
            vloc[name] = os.path.join(fvloc,fl)
        

    
    mykeys = vloc.keys()

    for key in mykeys:
        datamap[vloc[key]] = tloc[key] #data [vidofile] = lanefile





```

```python
def read_lane_lines(lane_file):
    lines = []
    with open(lane_file,"r") as f:
        lines = f.readlines()
        #print(len(lines))
        ilines = []
        for line in lines:
            points = line.strip().split()
            ipoints = []
            for i in range(0, len(points), 2):
                x = int(float(points[i]))
                y = int(float(points[i + 1]))
                ipoints.append([x,y])
            ilines.append(ipoints)



    return ilines
```

```python
y_col = np.linspace(400,580,19)
y_col #this is for prediction the lane curvature is 11 points
```

```python
data_line_map = {}
data_stline_map = {} #This is only used for bining the lines into different anchors for lane detection
data_coef_map  = {}
number_of_images = len(image_dict_keys)
for i in range(len(image_dict_keys)):
    image_loc = image_dict_keys[i]
    line_loc = datamap[image_loc]
    lines = read_lane_lines(line_loc)
    num_lines = len(lines)
    if(num_lines==0):
        continue #Lets skip these as some are not annoted even if they exist. We will use another model just figure out that
    new_lines = []
    new_st_lines= []
    new_coefs = []
    for line in lines:
        linarray = np.array(line)
        z=np.polyfit(linarray[:,1],linarray[:,0],3)
        zst = np.polyfit(linarray[:,1],linarray[:,0],1)#Just fit a straight line
        f = np.poly1d(z)
        fst = np.poly1d(zst)
        x_col = f(y_col)
        # if(max(x_col)>5000 or min(x_col)<-5000):
        #     continue #Discard this line as the line goes away
        new_lines.append(x_col) #only store the x, since y_col is fixed
        new_coefs.append(z)#append the coefficient that form the line
        new_st_lines.append(fst(y_col))

    
    if(len(new_lines)==0):
        continue

    data_line_map[image_loc] = new_lines.copy()
    data_coef_map[image_loc] = new_coefs.copy()
    data_stline_map[image_loc] = new_st_lines.copy()
    if(i%1000==0):
        print(i/number_of_images)
        
            
```
